```python
dataset2color = {}
dataset2color['train'] = '#FFA500' #'orange' 
dataset2color['burn'] = 'red' 
dataset2color['val'] = 'green' 

d_color = {}
d_color[0] = 'yellow' 
d_color[1] = "cyan"
d_color[2] = 'silver'
d_color[3] = "orange"


for category in df.data_category.unique():
    mask = df.data_category==category
    prices = df[mask].Close.values
    dates = df[mask].index.values
    regime = df[mask].regime.values
    
    plt.figure(figsize=(18,4))
    plt.plot(dates,prices,color=dataset2color[category], label=ticker)

    plt.title(f' Regime detection: {category}')
    plt.legend(loc=0)

    j = 0
    for i in range(1,len(prices)):
        if np.isnan(regime[i-1]):
            j=i
        elif (regime[i-1] == regime[i]) and (i< len(prices)-1):
            continue
        else:
            plt.axvspan(dates[j], dates[i], 
                       alpha=0.5, color=d_color[regime[i-1]], label="interval")
            j = i
    plt.show()
```

### Summarize 2-regime vs no-regime  (new logic: best train data rank +best val data rank+ lowest train-val)

```python
d_summary = {}
summarize = CROSSOVER.quick_compare_random
CROSSOVER.MAX_ATTEMPTS =2000
```

```python
## Expected run-time 90 mins for 27 tickers, 2000 attempts per ticker
for ticker in tickers:
    ## Fetch ticker data from disk
    print("-"*10, ticker,"-"*10)
    filename = f"data/{ticker}.pkl"
    df = pd.read_pickle(filename)
    df.sort_index(ascending=True, inplace=True)
    df.drop(["Dividends","Stock Splits"], axis=1, inplace=True)

    ## Add features
    df = add_data_splitter(df)
    df = add_features(df)
    df.dropna(inplace=True)

    ## Run PCA
    pca_cols = ['Volume', 'dir1d', 'high_low_diff','Open_prevClose_diff','ret3d','ret5d']
    X_pca = get_pca_components(df, pca_cols)

    ## Run HMM on train data
    train_mask = df.data_category=="train"
    n_regime_shifts, df['regime'] = run_hmm(X_pca, [.999,.95], train_mask)

    ## create cross_df
    cross_df = df[['Close','regime','data_category']].copy()
    cross_df.reset_index(inplace=True)

    ## Backtesting - 2Reg vs No-Reg    
    no_reg, dual_reg = summarize(cross_df, within=range(5,100),metrics = ['Sharpe'], moving_average='Simple', top_k=1)
    d_summary[ticker] = dual_reg.xs_Sharpe.val.values[0], no_reg.xs_Sharpe.val.values[0]
```

```python
xs_Sharpe_df = pd.DataFrame(d_summary, index=["Dual_reg", "No_reg"]).T
```

```python
xs_Sharpe_df.sort_values("Dual_reg")
```

```python
(xs_Sharpe_df.Dual_reg - xs_Sharpe_df.No_reg).mean()
```

```python
(xs_Sharpe_df.Dual_reg - xs_Sharpe_df.No_reg).hist(figsize=(8,4))
```

Clearly Dual-reg is better than no-reg by 0.12 Sharpe-Ratio difference
